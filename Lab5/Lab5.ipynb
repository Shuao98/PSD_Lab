Task 1: Write a Python program to draw a line with suitable label in the x axis, y axis and a title. The 
values of y should be twice of x. The range of x: 1 to 30. 

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(1,31)
y = x *2
plt.plot(x,y)
plt.xlabel('x', color='gray')
plt.ylabel('y', color='gray')
plt.title('My first plot',fontsize=20, fontname='Times New Roman')
```
